```python
"""
API Demonstration
Simple script to demonstrate API calling ability using the REST Countries API.

- Asks the user for a country name (e.g., "France", "Japan", "United States").
- Calls https://restcountries.com/v3.1/name/{country}
- Parses JSON response.
- Displays key information in a readable format.
"""

import urllib.request
import urllib.error
import json


def fetch_api_data(user_input):
    """Fetch country data from the REST Countries API based on user input.

    Args:
        user_input (str): Country name entered by the user.

    Returns:
        dict or None: Parsed country information (simplified), or None on error.
    """
    # Clean up input (strip spaces)
    country_name = user_input.strip()

    # Construct API URL (no API key needed)
    # 'fullText=false' makes the API more flexible with names
    url = f"https://restcountries.com/v3.1/name/{country_name}?fullText=false"

    try:
        with urllib.request.urlopen(url) as response:
            # Read and decode JSON payload
            data = json.loads(response.read().decode("utf-8"))

        # The API returns a list of matching countries; we take the first one
        if not data or not isinstance(data, list):
            print("No data found for that country.")
            return None

        country = data[0]

        # Extract relevant fields safely
        name_official = country.get("name", {}).get("official", "N/A")
        capitals = country.get("capital", [])
        capital = capitals[0] if capitals else "N/A"
        population = country.get("population", "N/A")
        region = country.get("region", "N/A")

        # Languages: stored as a dict of code -> name
        languages_dict = country.get("languages", {})
        languages = ", ".join(languages_dict.values()) if languages_dict else "N/A"

        result = {
            "official_name": name_official,
            "capital": capital,
            "population": population,
            "region": region,
            "languages": languages,
        }

        return result

    except urllib.error.HTTPError as e:
        # Commonly 404 when country not found
        print(f"HTTP Error: {e.code} - {e.reason}")
        return None
    except urllib.error.URLError as e:
        print(f"API Error (network problem): {e}")
        return None
    except (json.JSONDecodeError, KeyError, IndexError) as e:
        print(f"JSON parsing error: {e}")
        return None


def main():
    """Main function to demonstrate API calls."""
    print("=== API Demonstration: REST Countries ===\n")

    # Ask user for a country name
    user_input = input("Enter a country name (e.g., France, Japan, Brazil): ")

    # Call the API function
    result = fetch_api_data(user_input)

    # Display the results in a readable format
    if result:
        print("\n--- Country Information ---")
        print(f"Official Name : {result['official_name']}")
        print(f"Capital       : {result['capital']}")
        print(f"Population    : {result['population']}")
        print(f"Region        : {result['region']}")
        print(f"Languages     : {result['languages']}")
    else:
        print("\nNo country information could be retrieved.")

    print("\nAPI demonstration complete!")


if __name__ == "__main__":
    main()
```
